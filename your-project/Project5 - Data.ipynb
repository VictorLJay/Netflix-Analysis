```python
# Libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
import re
from datetime import datetime as dt
```

```python
# Importing Netflix DF
netflix = pd.read_csv("../data/Netflix_Titles.csv", )
netflix = netflix.drop(columns="description")
netflix["date_added"] = pd.to_datetime(netflix["date_added"])
netflix["date_added"] = netflix["date_added"].dt.strftime("%Y-%m")

# Importing IMDB DF
IMDB = pd.read_csv("../data/IMDB_results_nov-22-2020.csv", index_col=0)

# Joinig both tables
netflix = netflix.join(IMDB)

# Cleaning the DF mantaining the desired columns
netflix = netflix.drop(columns=["IMDB_titleID", "IMDB_title_name"])
```

##### INTENTO DE OBTENER TODOS LOS PAÍSES DE MANERA ÚNICA

Done

```python
#Obtención de todos los unique values. Ahora extracción de los mismos en otra cell
countries = list(netflix["country"].apply(str).unique())
countries = [x for x in countries if x != "nan"]
#len(countries) es 554
```

```python
only_first_country = []

for country in countries:
    only_first_country.extend(re.findall(r"^(.+?),", country))
```

```python
countries_clean = list(set(only_first_country))
countries_clean.sort()
print(countries_clean)
len(countries_clean)
```

```python
#ESTO FUNCIONA PARA UN PAÍS, REPLICARLO PARA MÁS
def dummy_country_genre(df, country):
    country_1 = df[df["country"].str.contains(country, case=False, regex=False, na=False)]
    
    return country_1[["listed_in"]].unstack().str.get_dummies(sep=", ").sum(level = 0).rename(index={"listed_in":country})

dummy_country_genre(netflix, "United States")
```

```python
# DEVUELVE TODO EL DF ONE HOT ENCODED

def dummy_country_genre_test2(df, country):
    list_of = []
    for c in country:
        country_1 = df[df["country"].str.contains(c, case=False, regex=False, na=False)]
        df2 = country_1[["listed_in"]].unstack().str.get_dummies(sep=", ").sum(level = 0).rename(index={"listed_in":c})
        list_of.append(df2)
          
    result = pd.concat(list_of).fillna(0).astype(int)
    
    return result

dummy_country_genre_test2(netflix, countries_clean)
```
